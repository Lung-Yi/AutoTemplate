```python
def read_num_lines(file_path):
    with open(file_path, "r") as f:
        data = f.readlines()
    return len(data)

before_list = []
after_list = []
nt_list = []

for rxn in list(all_rxn_class.keys()):
    before = read_num_lines(f"../data/{rxn}/MappingResult_{rxn}.txt")
    after = read_num_lines(f"../data/{rxn}/MappingResult_{rxn}.txt.processed")
    num_templates = read_num_lines(f"../data/{rxn}/all_templates_used.csv") - 1
    before_list.append(before)
    after_list.append(after)
    nt_list.append(num_templates)
```

```python
import numpy as np
# print the joint distribution of data before and after data-preprocess
def joint_data_distribution(all_rxn_class, before_list, after_list, nt_list, save_path):

    reaction_names = list(all_rxn_class.values())
    x = np.arange(len(all_rxn_class.keys()))
    width = 0.3
    fig, ax1 = plt.subplots(dpi=400)
    color1 = 'blue'
    ax1.bar(x, before_list, width, color=color1, label='Original reactions', alpha =0.4)
    ax1.bar(x, after_list, width, color=color1, label='Reactions after template-guided curation', alpha =1)
    ax1.set_ylabel('Number of reactions', color=color1, rotation=90, labelpad=15)
    ax1.tick_params(axis='y', labelcolor=color1, labelsize=8)
    
    ax2 = ax1.twinx()
    color2 = 'tab:red'
    ax2.set_ylabel('Number of generic templates', color=color2, rotation=-90, labelpad=15)
    ax2.bar(x + width, nt_list, width, color=color2, label='Generic templates', alpha =1)
    ax2.tick_params(axis='y', labelcolor=color2, labelsize=8)
    
    xticks = [z+width/2 for z in x]  # Calculate the center of bins
    ax1.set_xticks(xticks)
    ax1.set_xticklabels(reaction_names, rotation=-60, ha="left", fontsize=8)  # Set custom labels

    fig.tight_layout()
    ax1.legend(loc='upper left', fontsize=7)  # 將Data1的圖例放在左上角
    ax2.legend(loc='upper right', fontsize=7)
    plt.savefig(save_path, format="svg")
    plt.show()

                                             
joint_data_distribution(all_rxn_class, before_list, after_list, nt_list, save_path="../docs/bar_chart.svg")
print(nt_list)
```
