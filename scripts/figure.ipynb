```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})

all_rxn_class = {"AdamsDecarboxylation":"Adams decarboxylation",
                "Baylis-HillmanReaction":"Baylis-Hillman reaction",
                "Buchwald-HartwigCross-Coupling":"Buchwald-Hartwig cross coupling",
                "Chan_LamCoupling":"Chan-Lam coupling",
                "DielsAlder": "Diels-Alder",
                "FischerIndoleSynthesis":"Fischer indole synthesis",
                "Friedel-CraftsAcylation":"Friedel-Crafts acylation",
                "Friedel-CraftsAlkylation":"Friedel-Crafts alkylation",
                "GrignardReaction":"Grignard reaction",
                "HiyamaCoupling": "Hiyama coupling",
                "HuisgenCycloaddition":"Huisgen cycloaddition",
                "Hydrogenation":"Hydrogenation",
                "Kabachnik-FieldsReaction":"Kabachnik-Fields reaction",
                "KumadaCoupling":"Kumada coupling",
                "MannichReaction":"Mannich reaction",
                "NegishiCoupling":"Negishi coupling",
                "PausonKhandReaction":"Pauson-Khand reaction",
                "ReductiveAmination":"Reductive amination",
                # "SubstitutionReaction":"Substitution Reaction",
                "SuzukiCoupling":"Suzuki coupling",
                "WittigReaction":"Wittig reaction",
}
color_dict = {"no change": "dimgray", 
              "mapping curated": "blue",
              "reactant curated": "limegreen",
              "removed": "red"}
# color code:
# https://datascientyst.com/full-list-named-colors-pandas-python-matplotlib/
total_data = pd.read_csv("analyze_results.csv", index_col=0)
total_data = total_data.rename(all_rxn_class)
plot = total_data.plot(kind='barh',stacked=True, color= color_dict)
plt.legend(loc='upper left', prop = { "size": 7.5 })
plt.xlabel("Fraction (-)")

fig = plot.get_figure()
fig.savefig("output.svg", format="svg")

```
