```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_classes(data, threshold = 5, xlabel='', title = '', color = 'b', save_path=None):
    data = data[data["count"] > threshold]
    names, counts = data["template"], data["count"]
    # show = len(names)//6
    # names = [n.lower() if ((i+1)%show == 1) else '' for i, n in enumerate(names) ]
    num_labels = len(names)
    names = [1] + ['']*(num_labels - 2) + [num_labels]
    pos = np.arange(len(names))
    width = 1.0
    plt.figure(figsize=(6,5),dpi=100)
    ax = plt.axes()
    ax.set_xticks(pos + (width / 2))
    ax.set_xticklabels(names,)# rotation=-65, ha='left')
    plt.ylabel('Count (-)')
    plt.xlabel(xlabel)
    plt.title(title)
    plt.bar(pos, counts, width, color=color)
    if save_path:
        plt.savefig(save_path, format="svg", bbox_inches = "tight")

data_dir = "../data/"
reaction = "FischerIndoleSynthesis"
reaction_dir = os.path.join(data_dir, reaction)

rdchiral_data = pd.read_csv(os.path.join(reaction_dir, "rdchiral_templates.csv"))
universal_data = pd.read_csv(os.path.join(reaction_dir, "all_templates_used.csv"))
```

```python
plot_classes(rdchiral_data)
```

```python
plot_classes(universal_data)
```
